```python
import numpy as np
def kernel(x1, x2):
    sigma_f = 20
    sigma_l = 10
    return sigma_f**2 * np.exp(-1*(x1-x2)**2 / (2* sigma_l **2))

sigma_n = 0.05

x = np.array(list(range(1,11)))
y = np.array(list(range(2,21,2)))
k = np.zeros((10,1))
K = np.zeros((10,10))

target_point = 9.5
for i in range(10):
    for j in range(10):
        K[i,j] = kernel(x[i],x[j])
    k[i,0] = kernel(x[i],target_point)

t = k.T @ np.linalg.inv(K + sigma_n**2 * np.eye(10)) @ y.reshape([-1,1])

print(t)
```
